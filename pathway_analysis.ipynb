```python
model_name = 'GAT_Edge'
ss_path = path + '/Saliency/IG/CellLine/' + model_name + '/iqr_mean_baseline'

drug_dict, sal_dict =  make_gene_ss_dict(ss_path, type='drug')
norm_sal_dict = normalize_ss(sal_dict)
rank_dict = rank_ss(norm_sal_dict)
```
